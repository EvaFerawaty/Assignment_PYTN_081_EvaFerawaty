```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('london_crime_by_lsoa.csv')
```

```python
data.drop(['lsoa_code','minor_category','month'], axis=1, inplace=True)
data.head(4)
```

```python
crime2016 = data[(data['year']==2016)].groupby(['borough','major_category']).sum()['value'].sort_values(ascending=False)
crime2016.head(5)
```

```python
crime2016 = data.groupby(["major_category"])["value"].sum()
crime2016.plot(kind="pie", autopct="%.1f%%", explode=[0,0.25,0.25,1,0.75,0.5,0.25,0,0] ,legend=True, ylabel="", rot=10, figsize=(5,7))
plt.legend(bbox_to_anchor=(2.5, 1.0))
plt.title('London Crime 2016')
plt.axis('equal') 
plt.show()
```
